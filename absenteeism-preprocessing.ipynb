```python
import pandas as pd 
import numpy as np 
import matplotlib as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv("/Users/mohammadreza/Documents/myFiles/python/Untitled Folder/001 Absenteeism-data.csv")
raw_data
```

```python
data = raw_data.copy()
data= data.drop("ID",axis=1)
```

```python
pd.options.display.max_columns=None
pd.options.display.max_rows=None
data
```

```python
reason_columns = pd.get_dummies(data["Reason for Absence"], drop_first=True )
reason_columns
```

```python
reasont_1= reason_columns.loc[:,1:14].max(axis=1)
reasont_2= reason_columns.loc[:,15:17].max(axis=1)
reasont_3= reason_columns.loc[:,18:21].max(axis=1)
reasont_4= reason_columns.loc[:,22:].max(axis=1)
```

```python
data= pd.concat([data, reasont_1,reasont_2,reasont_3,reasont_4],axis=1)
data
```

```python
data= data.drop('Reason for Absence',axis=1)
```

```python
data.columns.values
```

```python
columns_names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'reason_1','reason_2','reason_3','reason_4']
data.columns = columns_names


```

```python
new_columns= ['reason_1','reason_2','reason_3','reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
data = data[new_columns]
```

```python
data_date= data.copy()
```

```python
data_date
```

```python
data_date["Date"] = pd.to_datetime(data_date["Date"],format='%d/%m/%Y')
data_date["Date"]
```

```python
month= []
```

```python
for i in range (data_date.shape[0]):
    month.append(data_date["Date"][i].month)
```

```python
month
```

```python
data_date["value of month"]= month
```

```python
dayValue  = []
```

```python
for i in range (data_date.shape[0]):
    dayValue.append(data_date["Date"][i].weekday())
```

```python
data_date["day value"]= dayValue
```

```python
data_date
```

```python
dataValues=['reason_1', 'reason_2', 'reason_3', 'reason_4', 'Date','value of month',
       'day value','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
data_date= data_date[dataValues]
```
